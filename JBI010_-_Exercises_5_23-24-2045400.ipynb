# Object-Oriented Card

Consider class `Card` (see below).

Add a method `glue` that takes another card as parameter,
and modifies `self` such that it keeps its own down side,
but copies the up side of the other card (as if the other card was glued on top).

**Notes**:
* Don't forget the implicit parameter `self`.
* Don't forget type hints.
  * N.B. The type `Card` must be written as a string: `'Card'`,
    because it is not yet defined when it is used.
* Add test cases to method `glue`.

````python
class Card:
    """
    A mutable card with an up and down side (non-empty strings).

    :attribute up: the up side of the card.
    :attribute down: the down side of the card.
    
    >>> Card('', 'O')
    Traceback (most recent call last):
        ...
    AssertionError: up and down must not be empty
    >>> card = Card('#', 'O')
    >>> card
    Card('#', 'O')
    >>> card.flip()
    >>> print(card)
    O (#)
    """
    
    def __init__(self, up: str, down: str) -> None:
        """
        Create card with given state.

        :param up: the up side of the card.
        :param down: the down side of the card.

        :return: a new card.
        """
        assert up and down, "up and down must not be empty"
        self.up = up
        self.down = down
        
    def __repr__(self) -> str:
        """
        Return machine-processible string representation of current state.

        :return: a string representation of the card.
        """
        return f"Card({self.up!r}, {self.down!r})"
    
    def __str__(self) -> str:
        """
        Return human-readable string representation of current state.

        :return: a string representation of the card.
        """
        return f"{self.up} ({self.down})"
    
    def flip(self) -> None:
        """
        Flip over this card.
        
        Modifies: self
        """
        self.up, self.down = self.down, self.up
        
#// BEGIN_TODO [glue] Glue card on top

    def glue(self, other: 'Card') -> None:
        """
        Glues another card on top, copying its up side and keeping its own down side.

        Args:
            other (Card): Another card to glue on top.
        """
        self.up = other.up
    
    def test_glue():
        card1 = Card('#', 'O')
        card2 = Card('X', 'Y')
        card1.glue(card2)
        assert card1.up == 'X' and card1.down == 'O'
        card2.glue(Card('A', 'B'))
        assert card2.up == 'A' and card2.down == 'Y'

#// END_TODO [glue]
````

# Points

We have a class ```Point``` that represents a point $(x, y)$.

````python
class Point:
    """
    Represents a point in 2D.
    """
    
    def __init__(self, x: float=0.0, y: float=0.0) -> None:
        """
        Initializes a `Point` object.
        :param x: coordinate x of the point
        :param y: coordinate y of the point
        """
        self.x = x
        self.y = y
    
    def __repr__(self) -> str:
        """
        Returns a string representation of a `Point` object.
        :returns: string representation of a `Point` object.
        """
        return f'Point({self.x}, {self.y})'



````

Create a function `compute_slope` that takes two `Points` and returns the slope of the line that exists between them. Then, create two points `point1` (3, 2) and `point2` (4, 8) to test the function. Notice that vertical lines do not have slopes. In these cases, you must return `None`. If the two points are the same, then there is no unique line between both points. Thus, in this case, you must also return `None`.

**Example:**
```python
# Input
compute_slope(point1, point2)

# Output 
6.0
```

````python
#// BEGIN_TODO [Points] Points 

def compute_slope(point1: Point, point2: Point) -> float:
    """
    Calculates the slope of the line passing through two points.

    Args:
        point1 (Point): First point.
        point2 (Point): Second point.

    Returns:
        float or None: Slope of the line. Returns None for vertical lines or identical points.
    """
    if point1.x == point2.x:
        return None  # Vertical line, slope is undefined
    else:
        slope = (point2.y - point1.y) / (point2.x - point1.x)
        return slope

point1 = Point(3, 2)
point2 = Point(4, 8)

#// END_TODO [Points]

compute_slope(point1, point2)
````

# Lines

Create a class `Line` that stores two `Points` (see exercise "Points") as attributes. It will also have its `slope` as an attribute. However, instead of receiving the slope as a parameter, you will compute the value by using the `compute_slope` function defined in the exercise "Points". Afterward, create the function `are_perpendicular`, which takes two `Line` objects as arguments and returns `True` if both lines are perpendicular,`False` otherwise.

**Notes:** 
- You will need to use the slope of the lines to get an answer. 
- Only consider lines with defined slopes, otherwise, return `False`.

**Example:**
```python
# Input 
line1 = Line(Point(-3, 2), Point(1, -2))
line2 = Line(Point(-1, -2), Point(3, 2))
are_perpendicular(line1, line2)

# Output
True
```

````python
#// BEGIN_TODO [Lines] Lines 
class Point:
    def __init__(self, x: float, y: float):
        """
        Represents a point in 2D space.

        Args:
            x (float): x-coordinate of the point.
            y (float): y-coordinate of the point.
        """
        self.x = x
        self.y = y

    def compute_slope(self, other_point: 'Point') -> float:
        """
        Computes the slope of the line passing through this point and another point.

        Args:
            other_point (Point): Another point through which the line passes.

        Returns:
            float: Slope of the line.
        """
        if self.x == other_point.x:
            raise ValueError("Slope is undefined for vertical lines.")
        return (other_point.y - self.y) / (other_point.x - self.x)

class Line:
    def __init__(self, point1: Point, point2: Point):
        """
        Represents a line segment in 2D space.

        Args:
            point1 (Point): First endpoint of the line segment.
            point2 (Point): Second endpoint of the line segment.
        """
        self.point1 = point1
        self.point2 = point2
        self.slope = point1.compute_slope(point2)

    def __repr__(self) -> str:
        """
        Returns a string representation of the 'Line' object.

        Returns:
            str: String representation of the 'Line' object.
        """
        return f"Line({self.point1}, {self.point2})"

def are_perpendicular(line1: Line, line2: Line) -> bool:
    """
    Checks if two lines are perpendicular.

    Args:
        line1 (Line): First line.
        line2 (Line): Second line.

    Returns:
        bool: True if lines are perpendicular, False otherwise.
    """
    try:
        
        return line1.slope * line2.slope == -1
    except ZeroDivisionError:
        return False  



#// END_TODO [Lines]

line1 = Line(Point(-3, 2),Point(1, -2))
line2 = Line(Point(-1, -2), Point(3, 2))

are_perpendicular(line1, line2)
````

# Line intersection

Create the function `compute_intersection()`, which takes two `Line` as arguments and returns the point where they intersect or `None` if there is no intersection. If both lines are the same, the function will also return `None`. Do not consider cases where the slope of at least one of the lines is undefined.

*Hint:* You can consider the linear equation $y = mx + b$, where $x$ and $y$ are the coordinates of a point in the line, $m$ is the slope of the line, and $b$ is the intercept. Computing the intercept first might also be a good idea.

**Example:**
```python
# Input 
line1 = Line(Point(-1, -1), Point(2, -2))
line2 = Line(Point(0, 1), Point(1, 0))

# Output
Point(0.5, 0.5)
```

````python
#// BEGIN_TODO [Line_intersection] Line intersection

class Point:
    def __init__(self, x: float, y: float):
        """
        Represents a point in 2D space.

        
        :param x: (float): x-coordinate of the point.
        :param y: (float): y-coordinate of the point.
        """
        self.x = x
        self.y = y

    def __repr__(self) -> str:
        """
        Returns a string representation of the 'Point' object.
        """
        return f'Point({self.x}, {self.y})'


class Line:
    def __init__(self, point1: Point, point2: Point):
        """
        Represents a line segment in 2D space.

    
        param point1 (Point): First endpoint of the line segment.
        param point2 (Point): Second endpoint of the line segment.
        """
        self.point1 = point1
        self.point2 = point2
        self.slope = (point2.y - point1.y) / (point2.x - point1.x)
        self.intercept = point1.y - self.slope * point1.x


def compute_intersection(line1: Line, line2: Line) -> Point:
    """
    Computes the intersection point of two non-parallel lines.

    
     :param line1: (Line): First line.
     :param line2: (Line): Second line.

   
     :returns 'Point': Intersection point of the two lines. Returns None if lines are parallel or identical.
    """
    if line1.slope == line2.slope:
        if line1.intercept == line2.intercept:
            return None  
        else:
            return None  
    x_intersection = (line2.intercept - line1.intercept) / (line1.slope - line2.slope)
    y_intersection = line1.slope * x_intersection + line1.intercept
    return Point(x_intersection, y_intersection)




#// END_TODO [Line_intersection]

line1 = Line(Point(-1, -1), Point(2, 2))
line2 = Line(Point(0, 1), Point(1, 0))

print(compute_intersection(line1, line2))
````

````python
#// BEGIN_TODO [Rolling_stones_albums] Rolling stones albums 

import csv

class Album:
    """
    This class represents a 'Album'.
    
    """
    def __init__(self, title: str = 'unknown', artist: str = 'unknown', genre: str = 'unknown', year: int = 0)-> None:
        """
        Initializes a 'Album' object.
        
        :Param name: is the name of the album.
        :Param artist: is the Artist of the album.
        :Param genre: is the genre of the album.
        :Param year: is the year the album is made.
        
        """
        self.title = title 
        self.artist = artist 
        self.genre = genre 
        self.year = year
    
    def __repr__(self)->str:
        """
        this represents a string notation of the 'Album'.
    
        """
        return "album({}, {}, {}, {})".format(self.title, self.artist, self.genre, self.year)
    
def read_album_dataset(file_path):
    albums = []
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  
        for row in reader:
            number, year, title, artist, genre, subgenre = row
            album = Album(title, artist, genre, int(year))
            albums.append(album)
    return albums
#// END_TODO [Rolling_stones_albums]

albums = read_album_dataset('datasets/albumlist.csv')
albums
````

# Most Popular Artist

Create the function `find_most_popular_artist`, which takes a list of `Albums` and returns the name of the artist that has more albums in the list. Use the `albums` variable declared in the exercise "Rolling Stones Albums" to test your output.

**Example:**
```python
# Input 
albums

# Output 
print(find_most_popular_artist(albums))
'The Beatles'
```

````python
#// BEGIN_TODO [Most_popular_artist] Most popular artist

from typing import List

class Album:
    def __init__(self, title: str, artist: str, genre: str, year: int):
        """
        Represents an 'Album' with its title, artist, genre, and release year.

        
        :param title (str): 'Album' title.
        :param artist (str): Artist name.
        :param genre (str): Genre of the album.
        :param year (int): Release year of the album.
        """
        self.title = title
        self.artist = artist
        self.genre = genre
        self.year = year

def find_most_popular_artist(albums: List[Album]) -> str:
    """
    Finds the most popular artist based on the number of albums in the list.

   
    :paramalbums (List[Album]): List of 'Album' objects.


    
    :returns str: Name of the most popular artist.
    """
    artist_counts = {}  
    for album in albums:
        artist = album.artist
        artist_counts[artist] = artist_counts.get(artist, 0) + 1
    
    
    most_popular_artist = max(artist_counts, key=artist_counts.get)
    return most_popular_artist



#// END_TODO [Most_popular_artist]

find_most_popular_artist(albums)
````
