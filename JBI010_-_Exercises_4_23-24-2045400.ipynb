# Graph Stored in a Dictionary

A (directed) _graph_ (or network) can be represented by a dictionary `d` of type `Dict[T, Set[T]]`,
  where `T` is a _hashable_ type, such as `int` or `str`:
* The _vertices_ of the graph (or nodes of the network) are stored as the keys: `d.keys()`.
* The _edges_ of the graph (or connections of the network) are pairs `(k, v)` where `v in d[k]`.  
  The dictionary maps each vertex `k` to the set `d[k]` of targets for edges with source `k`.

This is known as the **adjacency-list** representation.
(See e.g. [Representing graphs](https://www.khanacademy.org/computing/computer-science/algorithms/graph-representation/a/representing-graphs) on Khan Academy.)

Define the function `edge_count` that takes a graph `g: Dict[int, Set[int]]` as a parameter
and returns the number of edges in the graph.


**Notes**:
* Don't forget to write type hints and an appropriate docstring.
* Add _at least two_ relevant doctest examples:
  * one boundary case
  * one small typical case
* Use `g.values()` and `len`, and
  * either a `for`-loop and a `count` variable, or
  * `sum` and a list comprehension.
* N.B. Instead of `sum([item for item in iterable])`,
  it is more efficient to use `sum(item for item in iterable)`, because then the list of all items will _not_ be stored before summing.


**Example:**  
*Input:*  
<pre>
edge_count({1: {1, 2, 3}, 2: {1}, 3: set()})
</pre>

*Output:*  
<pre>
4
</pre>

````python
#// BEGIN_TODO [edge_count] Count edges in graph

from typing import Dict, Set

def edge_count(g: Dict[int, Set[int]]) -> int:
    """
    Count the number of edges in a graph represented as a dictionary.

    :param1:g (Dict[int, Set[int]]): A graph represented as a dictionary.

    Returns:int: The total number of edges in the graph.
    
    Examples:
    # Typical case: Graph with multiple vertices and edges
    >>> graph1 = {1: {1, 2, 3}, 2: {1}, 3: set()}
    >>> edge_count(graph1)
    4

    # Boundary case: Empty graph (no edges)
    >>> graph2 = {}
    >>> edge_count(graph2)
    0
    """
    # Use a list comprehension to get the sizes of all adjacency sets and sum them up
    edge_counts = [len(targets) for targets in g.values()]
    return sum(edge_counts)

#// END_TODO [edge_count]
````

# Remove Invalid Dates

Create the function `remove_invalid_dates`, which takes a list of tuples as an argument and returns a list of `datetime.date` objects with valid dates.
A valid year has a value equal to or less than 2022 (current year); a valid month is an integer between 1 and 12, and a valid day is an integer between 1 and 31.
Use comprehension to compute the expected list. 
You are also encouraged to use the `tuple_to_date` function developed in the exercise "Current Date".

Afterward, create a function `get_invalid_dates` that does the opposite of the `remove_invalid_dates` function: it also takes a list of tuples as an argument, but instead returns a list of tuple objects with invalid dates. Again, use comprehensions to compute the expected list.

**Note:** Preserve the lists order.

**Example:**
```python
# Program
dates = [(1903, 10, 3), (-230, -10, 2), (1947, 1, 2), (1990, 30, 30), (1678, 3, 30), 
         (680, 11, 23), (2500, 12, 12)]

print(remove_invalid_dates(dates))

# Output valid dates
[datetime.date(1903, 10, 3), 
 datetime.date(1947, 1, 2), 
 datetime.date(1678, 3, 30), 
 datetime.date(680, 11, 23)]

# Output invalid dates
[(-230, -10, 2), (1990, 30, 30), (2500, 12, 12)]

```

````python
#// BEGIN_TODO [Remove_invalid_dates] Remove invalid dates

# Function to remove invalid dates

from typing import Optional

def tuple_to_date(date_tuple: tuple[int, int, int]) -> Optional[datetime.date]:
    """
    Convert a tuple representing a date into a datetime.date object.
    
    :param date_tuple: A tuple with three integers representing year, month, and day.
    :return: A datetime.date object representing the given date, or None if the input is not valid.
    """
    year, month, day = date_tuple
    
    # Check for valid date components
    if 1 <= month <= 12 and 1 <= day <= 31 and year <= 2022:
        return datetime.date(year, month, day)
    else:
        return None
    
    
def remove_invalid_dates(dates: list[tuple[int, int, int]]) -> list[tuple[int, int, int]]:
    """
    Remove invalid dates from the given list of tuples.
    
    :param dates: A list of tuples representing dates.
    :return: A list of tuples with valid dates.
    """
    return [tuple_to_date(date) for date in dates if tuple_to_date(date)]

def get_invalid_dates(dates: list[tuple[int, int, int]]) -> list[tuple[int, int, int]]:
    """
    Get invalid dates from the given list of tuples.
    
    :param dates: A list of tuples representing dates.
    :return: A list of tuples with invalid dates.
    """
    return [date for date in dates if not tuple_to_date(date)]

#// END_TODO [Remove_invalid_dates]

dates = [(1903, 10, 3), (-230, -10, 2), (1947, 1, 2), (1990, 30, 30), (1678, 3, 30), 
         (680, 11, 23), (2500, 12, 12)]

print(remove_invalid_dates(dates))
print(get_invalid_dates(dates))
````

# Movie Dialogues

Create the function `extract_dialogue` which takes a line of dialogue (string) as input and returns a tuple with two elements: the name of the character (string) and its line in the dialogue (string). 
A typical input for the function looks like this: 

```python
"L872 +++$+++ u0 +++$+++ m0 +++$+++ BIANCA +++$+++ Okay -- you're gonna need to learn how to lie."
```

Use **regular expressions** to extract the name of the character and their line in the dialogue. 
Then, return a tuple like the following:

```python
("BIANCA", "Okay -- you're gonna need to learn how to lie.")
```

Notice that whitespaces at the beginning and end of both elements in the tuple have been removed.

Afterward, define the function `extract_all_lines` which gets a path to a file as input (string) and returns a list of tuples.
The function reads each line of the file and extracts the name of the character and its dialogue.
Each character-dialogue tuple is added to the list in the order in which each line is read from the file.
To do so, use the `extract_dialogue` function defined before.

**Example:**
```python
# Program
extract_all_lines('datasets/movie_lines.txt')

# Output
[('BIANCA', 'They do not!'),
 ('CAMERON', 'They do to!'),
 ('BIANCA', 'I hope so.'),

 ('BIANCA', "Let's go."),
 ('CAMERON', 'Wow'),

 ('CAMERON', 'No'),
 ...]
```

````python
#// BEGIN_TODO [Movie_dialogues] Movie dialogues
import re
from typing import List, Optional, Tuple

def extract_dialogue(line: str) -> Optional[Tuple[str, str]]:
    """
    Extracts character name and dialogue from a line of dialogue.
    
    Args:line (str) A line of dialogue in the specified format.
        
    Returns:Tuple[str, str] or None: A tuple containing character name and dialogue if a match is found,
            None otherwise.
    """
    match = re.match(r".+?\+\+\+\$\+\+\+(.+?)\+\+\+\$\+\+\+(.+)", line)
    return (match.group(1).strip(), match.group(2).strip()) if match else None

def extract_all_lines(file_path: str) -> List[Tuple[str, str]]:
    """
    Extracts all character dialogues from the given file.
    
    :param:file_path (str): Path to the file containing dialogues.
        
    :Returns:List[Tuple[str, str]]: A list of tuples containing character name and dialogue.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = [line.strip() for line in file.readlines()]
            return [extract_dialogue(line) for line in lines if extract_dialogue(line) is not None]
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return []




#// END_TODO [Movie_dialogues] 

extract_all_lines('datasets/movie_lines.txt')
````

# Movie Categories

Create the function `extract_categories`, which takes a line of a movie metadata file as input and returns a list of strings with the movie's categories. A typical input for the function looks like this:

```python
"m4 +++$+++ 48 hrs. +++$+++ 1982 +++$+++ 6.90 +++$+++ 22289 +++$+++ ['action', 'comedy', 'crime', 'drama', 'thriller']"
```

Use **regular expressions** to extract the categories from the input line. Remove whitespaces and quotations surrounding each category. Then, return a list like the following:

```python
['action', 'comedy', 'crime', 'drama', 'thriller']
```

Notice that whitespaces at the beginning and end of elements of the list, as well as extra quotes, have been removed.

Then, define the function `extract_all_categories`, which gets a path to a file as input (string) and returns a dictionary.
The function reads each line of the file and extracts the list of categories defined in such line.
To do so, use the `extract_categories` function defined before.
This list is then considered to update the dictionary of all categories, where keys are categories and values are the frequency of these categories in the movies.

**Example:**
```python
# Program
extract_all_categories('movie_titles_metadata.txt')

# Output
{'comedy': 21,
 'romance': 14,
 'adventure': 12,
 'biography': 5,
 ... }
```

````python
#// BEGIN_TODO [Movie_categories] Movie categories

import re
from collections import defaultdict
from typing import List, Dict

def extract_categories(line: str) -> List[str]:
    """
    Extracts categories from a line of movie metadata.
    
    :param1:line (str): A line of movie metadata in the specified format.
        
    :Returns:List[str]: A list of categories extracted from the input line.
    """
    match = re.match(r".+?\+\+\+\$\+\+\+.+?\+\+\+\$\+\+\+.+?\+\+\+\$\+\+\+.+?\+\+\+\$\+\+\+.+?\+\+\+\$\+\+\+(.+)", line)
    if match:
        categories_string = match.group(1).strip()
        categories_list = re.findall(r"'(.*?)'", categories_string)
        return [category.strip() for category in categories_list]
    else:
        return []

def extract_all_categories(file_path: str) -> Dict[str, int]:
    """
    Extracts and counts categories from a file containing movie metadata.
    
    :param1:file_path (str): Path to the file containing movie metadata.
        
    :Returns:Dict[str, int]: A dictionary where keys are categories and values are their frequencies in the movies.
    """
    categories_count = defaultdict(int)
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            for line in lines:
                categories = extract_categories(line)
                for category in categories:
                    categories_count[category] += 1
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    return categories_count


#// END_TODO [Movie_categories] 

extract_all_categories('datasets/movie_titles_metadata.txt')
````
